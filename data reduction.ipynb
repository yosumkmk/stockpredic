```python
import os
import pandas as pd
import numpy as np
import codecs
import yosuset
import glob
from mymodules.newaxis1 import *
from copy import deepcopy
import time
```

```python
%%time
years=list(map(str,list(range(1996,2015))))#2019
#初期コード
#新コード追加用
datafolder='./stdata'
os.makedirs(datafolder,exist_ok=True)
data1=np.zeros((10000,8,1),dtype=np.object)
data1[:,:,:]=np.nan
data=deepcopy(data1)

for year in years:
    foldername='./'+year+'data/unzip/'
    data=deepcopy(data1)
    for name in glob.glob(foldername+'T*.csv'):
        print(name)
        starttime=time.time()
        #0>日時、１＞株式コード、２＞取引場、４＞始値、５＞高値、６＞安値、７＞終値、８＞出来高
        darray=pd.read_csv(name,header=-1).as_matrix()
        #その日の株式コード一覧
        for i in range(len(darray[:,1])):
            try:
                darray1=darray[0:len(darray[:,1])-i,1].astype(np.float32)
                darray1=np.delete(darray1,np.where(np.isnan(darray1)),axis=0)
            except TypeError:
                pass
            except ValueError:
                pass
            else:
                darray=darray[0:len(darray1),:]
                darray[:,1]=darray1
                break
        data2=deepcopy(data1)
        for t in range(len(darray[:,1])):
            i=int(darray[t,1])
            data2[i,0:3,0]=darray[t,0:3]
            data2[i,3,0]=darray[t,8]
            data2[i,4:8,0]=darray[t,4:8]            
        data=np.concatenate([data,data2],axis=2)
        eltime=time.time()-starttime
        print(eltime)
    data=np.delete(data,[0],2)
    np.save((datafolder+'/stdata'+year),data)
```

```python
#旧バージョン
#遅い→arrayのサイズはあまり変えない方が早い、データサイズ大きくなると遅い→バッジ処理
years=list(map(str,list(range(1996,2019))))#2019
#初期コード
#新コード追加用
datafolder='./stdata'
os.makedirs(datafolder,exist_ok=True)

for year in years:
    foldername='./'+year+'data/unzip/'
    foldername='./'+year+'data/unzip/'
    name=glob.glob(foldername+'T*.csv')[0]
    darray=pd.read_csv(name,header=-1).as_matrix()
    stockcode=np.sort(list(yosuset.self(darray[:,1])))
    data=n001(np.array([darray[np.where(darray[:,1]==x)][np.argmax(darray[np.where(darray[:,1]==x)][:,8])] for x in stockcode]))
    for name in glob.glob(foldername+'T*.csv'):
        print(name)
        starttime=time.time()
        #0>日時、１＞株式コード、２＞取引場、４＞始値、５＞高値、６＞安値、７＞終値、８＞出来高
        darray=pd.read_csv(name,header=-1).as_matrix()
        #その日の株式コード一覧
        for i in range(len(darray[:,1])):
            try:
                darray1=darray[0:len(darray[:,1])-i,1].astype(np.float32)
                darray1=np.delete(darray1,np.where(np.isnan(darray1)),axis=0)
            except TypeError:
                pass
            except ValueError:
                pass
            else:
                darray=darray[0:len(darray1),:]
                darray[:,1]=darray1
                break
        scd=np.sort(list(yosuset.self(darray[:,1])))        
        #前日の株コード一覧
        stockcode1=deepcopy(stockcode)
        #前日までので足りないところを追加
        for t in list(set(scd)-set(stockcode1)):
            #コードリストに追加
            stockcode=np.append(stockcode,t)
            #追加株式コードを追加
            npn=np.zeros((1,data.shape[1],data.shape[2]))
            npn[npn==0]=np.nan
            npn[0,1,:]=t
            data=np.concatenate([data,npn],axis=0)
        data=data[np.argsort(stockcode,axis=0)]
        #当日の出来高最大のもの残す
        dsel=np.array([darray[np.where(darray[:,1]==x)][np.argmax(darray[np.where(darray[:,1]==x)][:,8])] for x in scd])        
        #足りない株コードを追加,並び替え
        scd1=deepcopy(scd)
        scd=np.append(scd,np.array(list(((set(stockcode)-set(scd))))))
        npn=np.zeros((dsel.shape[1]))
        npn[npn==0]=np.nan
        for t in list(set(stockcode)-set(scd1)):
            npn[1]=t
            dsel=np.concatenate([dsel,n10(npn)],axis=0)
        dsel=dsel[np.argsort(scd),:]
        #合体[code,種類、時間]
        data=np.concatenate([data,n001(dsel)],axis=2)
        eltime=time.time()-starttime
        print(eltime)
    data=np.delete(data,[0],2)
    np.save((datafolder+'/stdata'+year),data)
```
